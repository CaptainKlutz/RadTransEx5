```python
import numpy as np
import matplotlib.pyplot as plt
def findI1(path,temp):
    h=6.626E-34
    c=3E8
    l=1E-5
    kb=1.380E-23
    I=np.exp(-path)*((2*h*c*c)/(np.power(l,5)))*1/(np.exp((h*c)/(l*kb*temp))-1)
    return I
def findI2(path,temp):
    h=6.626E-34
    c=3E8
    l=1E-5
    kb=1.380E-23
    I=np.exp(-path)*(1-((2*h*c*c)/(np.power(l,5)))*1/(np.exp((h*c)/(l*kb*temp))-1))
    return I
spacing=500
n=1E16
H=8000
temp=[283,255]
pathRange=np.logspace(-24,-20,num=spacing)
I=np.ones(spacing)
fig,ax1=plt.subplots(figsize=(10,10))
for x in range (0,spacing):
    I[x]=findI1(n*H*pathRange[x],temp[0])
ax1.semilogy(n*H*pathRange,I,label="Temperature="+str(temp[0]))
for x in range (0,spacing):
    I[x]=findI2(n*H*pathRange[x],temp[1])
ax1.semilogy(n*H*pathRange,I,label="Temperature="+str(temp[1]))
ax1.grid(b=None,which="major",axis="both")
ax1.set_ylabel("ToA irradiance")
ax1.set_xlabel("Optical path")
ax1.legend()
```
